```python
import pandas as pd
# train test split
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from tensorflow.keras import layers
```

```python
# import data
test_data_file = './digit-recognizer/test.csv'
train_data_file = './digit-recognizer/train.csv'
test_data = pd.read_csv(test_data_file)
train_data = pd.read_csv(train_data_file)
```

```python
train_data.head()
```

```python
missing_values = [col for col in train_data.columns if train_data[col].isnull().any()]
print(missing_values)
```

```python
train_data.describe()
```

```python
y = train_data.label
X = train_data.drop('label', axis=1)
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

## Random forest

```python
# random forest 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix

randomForest = RandomForestClassifier(random_state=1, max_depth=10)

randomForest.fit(train_X, train_y)
```

```python
predictions = randomForest.predict(val_X)
accuracy = accuracy_score(val_y, predictions)
mae = mean_absolute_error(val_y, predictions)
cm = confusion_matrix(val_y, predictions)
print("The confusion matrix is: \n", cm)
print("The accuracy of the model is: ", accuracy*100, "%")
print("The mean absolute error of the model is: ", mae)
```

```python
# try the test data
predictTest = randomForest.predict(test_data)
print("predictTest: ", predictTest)
```

```python
# red neuronal
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
```

```python
inputSize = train_X.shape[1]
print("inputSize: ", inputSize)
batched_model = Sequential([
    Dense(64, input_shape=(inputSize,), activation="relu"),
    BatchNormalization(),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dense(64, activation='tanh'),
    BatchNormalization(),
    Dense(2, activation='softmax')
])
batched_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
